```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Paso 1: Obtener el contenido HTML de la página de Wikipedia
url = 'https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M'
response = requests.get(url)
html_content = response.content

# Paso 2: Analizar el HTML con BeautifulSoup y encontrar todas las tablas
soup = BeautifulSoup(html_content, 'html.parser')
tables = soup.find_all('table')

# Paso 3: Seleccionar la primera tabla (tabla 1)
if tables:
    table = tables[0]
    data = []
    for row in table.find_all('tr')[1:]:
        cols = row.find_all(['th', 'td'])  # Buscar tanto en celdas de encabezado como en celdas de datos
        postal_code = cols[0].text.strip()
        borough = cols[1].text.strip()
        neighborhood = cols[2].text.strip()
        
        # Agregar el código postal y el vecindario a la lista de datos
        data.append((postal_code, neighborhood))
    
    # Convertir la lista de datos en un DataFrame
    df = pd.DataFrame(data, columns=['CódigoPostal', 'Vecindario'])
    
    # Eliminar filas donde la columna "Vecindario" contiene "Not assigned"
    df = df[df['Vecindario'] != 'Not assigned']
    df = df[df['CódigoPostal'] != 'Not assigned']
    
    # Restablecer los índices del DataFrame
    df.reset_index(drop=True, inplace=True)
    
    # Mostrar las primeras filas del DataFrame resultante
    print("Tabla 1:")
    print(df)
    
else:
    print("No se encontraron tablas en la página.")

```

```python
import pandas as pd

#Unifiación de listas para trabajar con una y luego separar

# Crear una lista con todos los datos de ambas columnas
datos_combinados = df['CódigoPostal'].tolist() + df['Vecindario'].tolist()

# Crear un nuevo DataFrame con una sola columna
nuevo_df = pd.DataFrame({'Código': datos_combinados})

# Imprimir el nuevo DataFrame
print(nuevo_df)

```

```python
# Limpiando indices corruptos No Asignados
# Índices a eliminar
indices_a_eliminar = [17, 18, 21, 22, 29, 30, 31, 32, 33, 34, 35, 36, 37]

# Eliminar las filas correspondientes a los índices especificados
nuevo_df = nuevo_df.drop(indices_a_eliminar).reset_index(drop=True)

# Ordenar el DataFrame por la columna "Código"
nuevo_df = nuevo_df.sort_values(by='Código')

# Alinear todos los datos en la columna "Código" a la izquierda
nuevo_df['Código'] = nuevo_df['Código'].str.lstrip()


# Imprimir el DataFrame resultante
print(nuevo_df)


```

```python
import pandas as pd

# Tabular todos los datos hacia la izquierda eliminando los espacios en blanco después del código postal
nuevo_df['Código'] = nuevo_df['Código'].str.replace(r'\s+', '')

# Identificar los datos que siguen el patrón estándar de código postal y vecindario entre paréntesis
patron_estandar = nuevo_df['Código'].str.contains(r'\(\w+\s*\/\s*\w+\)')

# Separar los datos que siguen el patrón estándar
nuevo_df.loc[patron_estandar, ['Código Postal', 'Vecindario']] = nuevo_df.loc[patron_estandar, 'Código'].str.extract(r'(\w+\d+\w+)\s*([^()]*)\(([^()]*)\)')

# Identificar los datos que no siguen el patrón estándar y dividirlos manualmente
no_estandar = ~patron_estandar
nuevo_df.loc[no_estandar, 'Código Postal'] = nuevo_df.loc[no_estandar, 'Código'].str.extract(r'(\w+\d+\w+)')[0]
nuevo_df.loc[no_estandar, 'Vecindario'] = nuevo_df.loc[no_estandar, 'Código'].str.extract(r'\(([^()]*)\)')[0]

# Imprimir el DataFrame resultante
print(nuevo_df)
```

```python
#Se crea dataFrame con cordenadas
import pandas as pd

# Cargar el archivo CSV en un DataFrame
ruta_archivo = 'C:/Users/cl164575616/Documents/Geospatial_Coordinates.csv'
df_cordenadas = pd.read_csv(ruta_archivo)

# Mostrar las primeras filas del DataFrame
print(df_cordenadas.head())
print(df_cordenadas.columns)
```

```python
import pandas as pd

# Función para buscar las coordenadas por código postal
def buscar_coordenadas(codigo_postal):
    resultado = df_cordenadas[df_cordenadas['Postal Code'] == codigo_postal]
    if not resultado.empty:
        latitud = resultado.iloc[0]['Latitude']
        longitud = resultado.iloc[0]['Longitude']
        return latitud, longitud
    else:
        return None, None

# Buscar y agregar las coordenadas al DataFrame df_final
df_final[['Latitude', 'Longitude']] = df_final['CódigoPostal'].apply(buscar_coordenadas).apply(pd.Series)

# Mostrar el DataFrame df_final con las nuevas columnas Latitude y Longitude
print(df_final)
```

```python
import folium

#Había problemas con los nombres y las cordenadas, así que se saco el promedio y los carácteres especiales de los nombres

# Calcular la ubicación promedio para centrar el mapa
latitud_promedio = df_final['Latitude'].mean()
longitud_promedio = df_final['Longitude'].mean()

# Crear un mapa centrado en la ubicación promedio
mapa = folium.Map(location=[latitud_promedio, longitud_promedio], zoom_start=10)

# Limpiar los nombres de los vecindarios para eliminar caracteres especiales
df_final['Vecindario'] = df_final['Vecindario'].str.replace('[^a-zA-Z0-9\s]', '', regex=True)

# Añadir marcadores para cada vecindario
for index, row in df_final.iterrows():
    folium.Marker([row['Latitude'], row['Longitude']], popup=row['Vecindario']).add_to(mapa)

# Mostrar el mapa
mapa
```
